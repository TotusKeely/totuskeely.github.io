Use the Maclaurin series expansion for sin x to estimate sin(π/3)
Set the maximum number of iteration to maxit = 100 and use the tolerance
(stopping criterion) es = 10^−6. You may use MATLAB, Python, C, or any
other programming language.

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Establish the variables
x = (np.pi/3)
es = 10**(-6)
val = 0
#make en large initially to ensure the while loop runs
en = 10**8
maxit = 100
#iteration counter
n = 0
enVals = []

while en >= es and n < maxit:
    nterm_x = ((-1)**n)*(x**(2*n+1))/(math.factorial(2*n+1))
    val = val + nterm_x
    en = abs(math.sin(x) - val) 
    n = n + 1
    #add val to an array for the graph
    enVals.append(en)

print("Estimated Value: ", val)
print("Calculated Value: ", math.sin(x))
print("Number of iterations ran: ", n)
print("en at",n, "iterations: ", en) 

n_vals = np.arange(1, n+1)
plt.plot(n_vals, enVals)
plt.xlabel('iterations')
plt.ylabel('Error for each iteration')
plt.title ('itearation vs. en')
```
